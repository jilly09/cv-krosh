```python
import subprocess
import logging
from datetime import datetime

# 1. ĞĞ°ÑÑ‚Ñ€Ğ¾Ğ¹ĞºĞ° Ğ»Ğ¾Ğ³Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ñ
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(message)s",
    handlers=[
        logging.FileHandler(f"train_sdxl_{datetime.now().strftime('%Y%m%d_%H%M%S')}.log"),
        logging.StreamHandler()
    ]
)

# 2. ĞšĞ¾Ğ½Ñ„Ğ¸Ğ³ÑƒÑ€Ğ°Ñ†Ğ¸Ñ ĞºĞ¾Ğ¼Ğ°Ğ½Ğ´Ñ‹
command = [
    "/home/jupyter/.local/bin/accelerate", "launch",
    "diffusers/examples/text_to_image/train_text_to_image_lora_sdxl.py",
    "--pretrained_model_name_or_path=stabilityai/stable-diffusion-xl-base-1.0",
    "--train_data_dir=altair_dataset",
    "--resolution=1024", "--random_flip",
    "--train_batch_size=1",
    "--num_train_epochs=100",
    "--use_8bit_adam",
    "--learning_rate=1e-06", "--lr_scheduler=constant", "--lr_warmup_steps=0",
    "--mixed_precision=no",
    "--output_dir=mymodel4",
    "--checkpointing_steps=600"
]

# 3. Ğ—Ğ°Ğ¿ÑƒÑĞº Ğ¿Ñ€Ğ¾Ñ†ĞµÑÑĞ° Ñ Ğ»Ğ¾Ğ³Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸ĞµĞ¼
try:
    logging.info("ğŸš€ Starting SDXL training...")
    logging.info("Command: " + " ".join(command))
    
    process = subprocess.Popen(
        command,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        text=True,
        bufsize=1,
        universal_newlines=True
    )
    
    # Ğ’Ñ‹Ğ²Ğ¾Ğ´ Ğ»Ğ¾Ğ³Ğ¾Ğ² Ğ² Ñ€ĞµĞ°Ğ»ÑŒĞ½Ğ¾Ğ¼ Ğ²Ñ€ĞµĞ¼ĞµĞ½Ğ¸
    for line in iter(process.stdout.readline, ''):
        logging.info(line.strip())
    
    process.wait()
    logging.info("âœ… Training completed successfully!")
    
except KeyboardInterrupt:
    logging.warning("âš ï¸ Training interrupted by user!")
    process.terminate()
except Exception as e:
    logging.error(f"âŒ Training failed: {str(e)}", exc_info=True)
finally:
    if process.poll() is None:
        process.kill()
    logging.info("ğŸ§¹ Cleanup completed.")
```
